```python
import pandas as pd
```

```python
data_penjualan = pd.read_csv("data_penjualan.csv")
nama_barang = pd.read_csv("nama_barang.csv")
```

## 1. merge

```python
## merge menggabungkan 2 dataframe secara horizontal dan di hubungkan dengan nilai yang sama dengan 2 column yanbg sama sebgai key
```

```python
hasil = pd.merge(nama_barang, data_penjualan, on="fid", how="right")
```

```python
hasil
```

```python
""" method ada beberapa seperti sql
1. inner = menampilkan data dengan yang sama dengan 
2. right = menampilkan semua dataframe di kanan dan menampilkan data yang sama dengan relasi, jika tidak akan menampilkan nilai NaN
3. left = hampir sama dengan dengan right """
```

## 2. join

```python
## join hampir sama dengan merge, yang jhadi pembeda adlah lebih mudah dengan index sebagai key, jadi tidak perlu mendifinisaka on
```

```python
coba = nama_barang.join(data_penjualan, how="inner")
```

```python
## jadi index harus sama
```

## 3. concatenate atau concat()

```python
## menggabungkan 2 nilai dengann vertical dan hampir sama denhan UNION di sql
```

```python
data1 = pd.DataFrame({
    "A": [1,2,3,4,5],
    "B": [6,7,8,9,10]
})
data2 = pd.DataFrame({
    "A": [11,12,13,14,15],
    "B": [16,17,18,19,20]
})

merged = pd.concat([data1,data2], axis=1, )
merged
```

## 4. Grouping

```python
# menggabungkan value yang memiliki nama yang sama berdasrakan kolumn tertentu
```

```python
lepi = pd.read_csv("laptop_price_dataset.csv")
```

```python
lepi
```

```python
dataNegara = pd.DataFrame({
    "nama": ["Gilbrant", "Roychan", "Kholbi", "Cuy"],
    "negara":["Indonesia", "Chine", "Indonesia", "Chine"]
    
})

dataNegara.groupby(["negara"])["negara"].count()
```

```python
lepi.groupby(["Company","GPU_Company"])["Inches"].mean().sort_values(ascending= False)
```

## aggregasi function memngkinkan kita melakuka perhitungan lebih dari sekali

```python
lepi.groupby(["Company"]).agg({"Price (Euro)": ["mean","min","max"]})
```

```python
## di sini bisa di lihat bahwa untuk kolomn Euro di lakukah 3 kali perhitungan
```

## dan aggregasi ini bs di lakukan lebih dari 1 kolom  

```python
lepi.groupby(["Company", "GPU_Company"]).agg({"RAM (GB)": ["mean", "min", "max"],
                                              "Price (Euro)": ["mean", "min", "max"]
 }).sort_values(by=("RAM (GB)", "mean"), ascending= False)
```

```python
## dan bs di lihat di atas untuk sort_values bs di lakukan dengan mentarget agg dan kolom
```

```python
lepi.groupby(["GPU_Company"]).agg({"RAM (GB)": "max"}).sort_values(by= "RAM (GB)", ascending=False)
```

## Pivot Tabel

```python
## pivot table memungkinkan kita merekab dan meng aggregasi data
```

```python
result=lepi.pivot_table(index="Company", columns= ["RAM (GB)","GPU_Company"], values="Product", aggfunc="count", fill_value = 0 ).sort_values(by=(2, "Intel")).astype(int)
#result.to_csv("hasil_pivot.csv")
result
```
